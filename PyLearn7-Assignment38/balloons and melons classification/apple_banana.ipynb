```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
balloon_length = np.random.normal(30, 0.5, 400)
balloon_width = np.random.normal(40, 0.5, 400)
```

```python
melon_length = np.random.normal(20, 0.5, 400)
melon_width = np.random.normal(15, 0.5, 400)
```

```python
plt.scatter(balloon_length, balloon_width, color="red")
plt.scatter(melon_length, melon_width, color="yellow")
plt.show()
```
