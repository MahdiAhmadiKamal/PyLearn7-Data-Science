```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from bidi.algorithm import get_display
from arabic_reshaper import reshape
```

```python
data = pd.read_csv("digikala-orders.csv")
data.head()
```

```python
data['DateTime_CartFinalize'] = pd.to_datetime(data['DateTime_CartFinalize'])

data['OrderYearMonth'] = data['DateTime_CartFinalize'].map(lambda date: 100*date.year + date.month)

data.head()
```

```python
data['Orders'] = data['ID_Order']
data_1 = data.groupby(['OrderYearMonth'])['Orders'].count().reset_index()
data_1
```

```python
plt.figure(figsize=(24,6))
plt.plot(data_1['OrderYearMonth'].to_numpy(dtype=str), data_1['Orders'], marker='o')
plt.grid()
plt.xticks(rotation=45)
plt.xlabel("year & month")
plt.ylabel("orders")
plt.title("Orders count")
plt.savefig("output/1.jpg")
plt.show()
```

```python
data_city = data.groupby(['city_name_fa'])['ID_Customer'].nunique().reset_index()
data_city.head()
```

```python
result = data_city.sort_values('ID_Customer', ascending=False)[:8]
s = data_city.sort_values('ID_Customer', ascending=False)[8:]['ID_Customer'].sum()
result.loc[len(result)]=['Others', s]
result.head(10)
```

```python
persian_labels = [get_display(reshape(label)) for label in result['city_name_fa'].to_numpy(dtype='str')]

plt.figure(figsize=(12, 12))
plt.pie(result['ID_Customer'], labels=persian_labels, autopct='%1.1f%%')
plt.savefig("output/2.jpg")
plt.show()
```
